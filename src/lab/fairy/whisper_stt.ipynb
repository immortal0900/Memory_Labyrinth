```python
import whisper
model = whisper.load_model("large")  # 로드될 때 디바이스가 결정됨
```

```python
# 위스퍼 정식 
import runpod

import torch
sound_list = [
    "record.wav",
    "record1.wav",
    "record2.wav",
    "record3.wav",
    "record4.wav",
    "record5.wav"
]
print("CUDA available:", torch.cuda.is_available())
def handler(event):
    """
    event = {
        "input": {
            "audio_path": "record.wav"
        }
    }
    """
    audio_path = event["input"]["audio_path"]

    result = model.transcribe(
        audio_path,
        language="ko"
    )

    return {
        "text": result["text"]
    }

# RunPod serverless 시작
runpod.serverless.start({
    "handler": handler
})

print("torch cuda available:", torch.cuda.is_available())
print("torch cuda device:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else None)
for sound in sound_list:
    result = model.transcribe(sound_list[0], language="ko")
    print(result["text"])
```

```python
import sounddevice as sd
import soundfile as sf

def record_wav(path, seconds, sr=16000):
    print(f"Recording {seconds}s @ {sr}Hz ...")
    audio = sd.rec(int(seconds * sr), samplerate=sr, channels=1, dtype="int16")
    sd.wait()
    sf.write(path, audio, sr, subtype="PCM_16")
    print("Saved:", path)

record_wav("record_test02.wav", seconds=3, sr=16000)
```
