```python
import requests

API_URL = "https://clovaspeech-gw.ncloud.com/recog/v1/stt"
API_KEY = "YOUR_CLOVASPEECH_SECRET_KEY"
AUDIO_PATH = "record.wav"

params = {
    "lang": "Kor",   # Kor | Eng | Jpn | Chn
}

headers = {
    "Content-Type": "application/octet-stream",
    "X-CLOVASPEECH-API-KEY": API_KEY,
}

with open(AUDIO_PATH, "rb") as f:
    response = requests.post(
        API_URL,
        params=params,
        headers=headers,
        data=f
    )

print(response.status_code)
print(response.json())
```
